```python
def calculate_accuracy(actual, expected):
  """Calculates the accuracy of predicted values against expected values.

  Args:
    actual: A list of actual values.
    expected: A list of expected values.

  Returns:
    The accuracy as a percentage.
  """
  if not actual or not expected or len(actual) != len(expected):
    return 0.0

  correct_count = 0
  for i in range(len(actual)):
    if actual[i] == expected[i]:
      correct_count += 1

  return (correct_count / len(actual)) * 100


accuracy_one = calculate_accuracy(actual_one, expected_one)
print(f"Accuracy for actual_one: {accuracy_one:.2f}%")

accuracy_two = calculate_accuracy(actual_two, expected_two)
print(f"Accuracy for actual_two: {accuracy_two:.2f}%")

accuracy_three = calculate_accuracy(actual_three, expected_three)
print(f"Accuracy for actual_three: {accuracy_three:.2f}%")

accuracy_four = calculate_accuracy(actual_four, expected_four)
print(f"Accuracy for actual_four: {accuracy_four:.2f}%")
```
